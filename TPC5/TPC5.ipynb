```python
#TPC5

import time
cinema=[]

def inserirSala(cinema):
    filme=input("Introduza o nome do filme")
    for s in cinema:
        nlugares,vendidos,nome=s
        if nome == filme:
            print("O fime já existe")
        menu()
    nlugares = int(input("Introduza o número de lugares sa sala"))
    vendidos=[]
    cinema.append(nlugares,vendidos,filme)
    print("Sala criada com sucesso")
    print("Salas:",cinema)
    menu()

def removerSala(cinema):
    index = int(input("Introduza o índice da sala"))
    cinema.pop(index)
    print("Sala removida com sucesso")
    print("Salas:", cinema)
    menu()

#listar salas
def listarSalas(cinema):
    print("Sala  Filme")
    print("-------")
    for p in cinema:
        nlugares,vendidos,filme=p
        print("Sala# {0}".format(cinema.index(p))," ",filme)
    time.sleep(1)
    menu()
#lugares disponiveis
def lugarDisponivel(cinema,filme,lugar):
    for p in cinema:
        nlugares,vendidos,nome=p
        if filme == nome:
            if lugar in vendidos:
                res=False
            else:
                res=True
        else:
            res=False
    return res
def vendeBilhete(cinema,filme,lugar):
    if lugarDisponivel(cinema,filme,lugar):
        for p in cinema:
            nlugares,vendidos,nome=p
            if filme == nome:
                vendidos.append(lugar)
        print("Lugar vendido com sucesso")
        print("Salas:", cinema)
    else:
        print("Lugar indisponível")
    menu()
#lista de disponibilidade de sala
def listarDisponibilidadeSala(cinema):
    print("Sala  Filme  Lugares Disponíveis")
    print("--------")
    for p in cinema:
        nlugares,vendidos,filme=p
        print("Sala#{0}".format(cinema.index(p))," ",filme," ",nlugares - len(vendidos))
    time.sleep(1)
    menu()
def menu():
    print("""
    ------------------------
    (1) Criar Sala
    (2) Remover Sala
    (3) Listar Salas
    (4) Lugares Disponíveis
    (5) Vender Bilhetes
    (6) Consultar Salas
    (0) Sair/Reset
    -----------------------
    """)
    menuOption=int(input("Escolha uma função:"))
    if menuOption==1:
        inserirSala(cinema)
    elif menuOption==2:
        removerSala(cinema)
    elif menuOption==3:
        listarSalas(cinema)
    elif menuOption==4:
        filme=input("Introduza o nome do filme")
        lugar=int(input("Introduza o lugar"))
        print("Lugar disponível?",lugarDisponivel(cinema,filme,lugar))
    elif menuOption==5:
        filme=input("Introduza o nome do filme")
        lugar=int(input("Introduza o lugar"))
        vendeBilhete(cinema,filme,lugar)
    elif menuOption==0:
        print("Reset efetuado com sucesso! Muito Obrigada")
        print("Sala:",cinema)
menu()

   




    
```
